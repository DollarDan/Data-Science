```python
def plot(list_of_vectors):
    i = 0
    colors =['r','g','b', 'y']
    for vec in list_of_vectors:
        plt.quiver(vec[0], vec[1], vec[2], vec[3], scale_units = 'xy', angles = 'xy', scale=1, color = colors[i%len(colors)])
        i += 1
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
        
```

```python
plot([(0,0,2,3), (0,0,-4,5),(0,0,6,2),(0,0,-3,-5)])
```

# vector Dot Product

```python
a = np.asarray([3, 4])
b = np.asarray([4, -1])
```

Dot Product formula is $\vec{a}\cdot\vec{b} = |\vec{a}||\vec{b}| \cos(\theta) = a_x b_x + a_y  b_y$

```python
a_dot_b = np.dot(a,b)
```

```python
a_dot_b
```

# Vector Projection : $a_b = |\vec{a}| \cos(\theta) = |\vec{a}|\frac{\vec{a}\cdot\vec{b}}{|\vec{a}||\vec{b}|} = \frac{\vec{a}\cdot\vec{b}}{|\vec{b}|}$

```python
a_b = np.dot(a,b)/np.linalg.norm(b)
```

```python
a_b
```

# Unit Vector : $ \vec{a_b} = a_b\hat{b} = a_b \frac{\vec{b}}{|\vec{b}|} $

```python
vec_a_b = (a_b/ np.linalg.norm(b))*b

print(vec_a_b)
```

```python
# the sbove is the projection of vector a on vector b.
# lets plot b and vec_a_b
# a = [3, 4]
# b = [4, -1]
# vec_a_b = [ 1.88235294 -0.47058824]
```

```python
plot([np.asarray([0,0,3,4]), np.asarray([0,0,4,-1]), np.asarray([0,0, 1.8823294, -0.47058824])])
```

```python
# you can see a in red, b in green and vec_a_b in blue
```

```python
# now lets see how vec_b_a looks like
# vec_a_b = (a_b/ np.linalg.norm(b))*b
b_a =  np.dot(b,a)/np.linalg.norm(a)
vec_b_a = (b_a/ np.linalg.norm(a))*a 

vec_b_a
```
