# Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.random((10,1))
```

```python
x
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(x, '*')
plt.show()
```

```python
plt.plot(x, '*-')
plt.show()
```

```python
x = np.linspace(0,10, 100)
y = np.power(x, 0.5)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, '+')
plt.show()
```

```python
x = np.linspace(0,100, 50)
y = np.power(x, 2)
plt.plot(x,y, '+')
plt.show()
```
